# The Oscars

This Project is about visualization. i selected the dataset from project 4(major league).

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import datetime as dt
import seaborn as sns
```

```python
spi_matches = pd.read_csv('https://raw.githubusercontent.com/aminshojaei/The-Oscars/master/dataset/spi_matches.csv', low_memory = False)
```

```python
spi_matches
```

- As it is obvious, there are some NaN cells. In hence at first I will go for cleaning the dataset of NaN values.

```python
spi_matches=spi_matches.dropna(how='any')      # for removing NaN data rows
spi_matches.isnull()
```

```python
spi_matches.info()
```

- Now the dataset is clear from any NaN value. In next step I want to divid dataset based on year and month. In hence, I wanna add to columns. One for year and another for month which show that which year and month are two teams are playing.
- for doing this, it is neede to change the type of the date column to datetime. So following code are provided.

```python
spi_matches['date']=spi_matches['date'].apply(lambda x: dt.datetime.strptime(x,'%Y-%m-%d'))
spi_matches.head()
```

```python
spi_matches.dtypes
```

- So the type of the date is changed to datetime. Now I wanna add two columns as year and month. Then I wanna drop those columns which are irrelevent to visualization.

```python
spi_matches['year_of_date'] = spi_matches['date'].dt.year
spi_matches['month_of_date'] = spi_matches['date'].dt.month
dataset=spi_matches.copy()
spi_matches =spi_matches.drop(['league_id','importance1', 'importance2','xg1','xg2','nsxg1','nsxg2'], axis = 1)
spi_matches.drop_duplicates()
spi_matches.head()
```

## Visualization

### Heatmap

If you have a data set with many columns, a good way to quickly check correlations among columns is by visualizing the correlation matrix as a heatmap.A heatmap is a two-dimensional graphical representation of data where the individual values that are contained in a matrix are represented as colors. 

```python
fig, ax = plt.subplots(figsize=(15,15)) 
cor = spi_matches.corr()
sns.heatmap(cor, square = True, linewidths=0.1, annot=True, ax=ax)
plt.show()
```

- From above map, you can see that some columns are more correlated. for example, prob1 & proj_score1 or prob2 & proj_score2. In addition, adj_score1 & score1 or adj_score2 & score2 are more correlated too.

### Pairplot

- A pairs plot allows us to see both distribution of single variables and relationships between two variables. Pair plots are a great method to identify trends for follow-up analysis

```python
sns.pairplot(spi_matches, vars=['spi1','spi2','prob1','prob1','probtie','proj_score1','proj_score2','prob1','prob2','score1','score2'])
```

### Matches in each League

- Now I wanna visualize more detail from dataset. Now I want to know that how many play was in each year and month based on our dataset.

```python
league=spi_matches['league'].value_counts().to_frame()
S=pd.Series(list(np.arange(0,len(league),1)))
league['Play'] = league.index.tolist()
league.rename(columns={'league':'number of matches', 'Play':'League'}, inplace=True)
cols=league.columns.tolist()
cols=cols[-1:]+cols[:-1]
league = league[cols]
league=league.set_index([S])
league.head()
```

```python
plt.figure(figsize=(15,10))
sns.set(style='whitegrid')
ax = sns.barplot(x='number of matches',y='League',data=league)
ax.set(xlabel='number of matches', ylabel='League')
plt.show()
```

### Matches by Year & Month

```python
year=spi_matches['year_of_date'].value_counts().to_frame()
S=pd.Series(list(np.arange(0,len(year),1)))
year['Play'] = year.index.tolist()
year.rename(columns={'Play':'year', 'year_of_date':'number of matches'}, inplace=True)
cols=year.columns.tolist()
cols=cols[-1:]+cols[:-1]
year = year[cols]
year=year.set_index([S])

plt.figure(figsize=(15,10))
sns.set(style='whitegrid')
ax = sns.barplot(x='year',y='number of matches',data=year)
ax.set(xlabel='number of matches', ylabel='League')
plt.title("Matches by year", fontsize=20)
plt.show()
```

```python
month=spi_matches['month_of_date'].value_counts().to_frame().sort_values(by='month_of_date')
S=pd.Series(list(np.arange(0,len(month),1)))
month['Play'] = month.index.tolist()
month.rename(columns={'Play':'month', 'month_of_date':'number of matches'}, inplace=True)
cols=month.columns.tolist()
cols=cols[-1:]+cols[:-1]
month = month[cols]
month=month.set_index([S]).sort_values(by='month')

plt.figure(figsize=(15,10))
sns.set(style='whitegrid')
ax = sns.barplot(x='month',y='number of matches',data=month)
ax.set(xlabel='number of matches', ylabel='League')
plt.title("Matches by month", fontsize=20)
plt.show()
```

- Above graph shows that number of plays in june and july is at the lowest rate. 

### Number of Goals 

```python
spi_matches["total_goal"] = spi_matches["score1"]+spi_matches["score2"]
sum_year = spi_matches.groupby("year_of_date").agg({"total_goal":"sum"})
sum_month = spi_matches.groupby("month_of_date").agg({"total_goal":"sum"})
mean_year = spi_matches.groupby("year_of_date").agg({"total_goal":"mean"})
mean_month = spi_matches.groupby("month_of_date").agg({"total_goal":"mean"})
sum_year
fig, ax = plt.subplots(4 , figsize=(18,15))
fig.suptitle("Number of total goal by year & month")
ax[0].plot(sum_year, color='green',label='Normal')
ax.flat[0].set(xlabel='year',ylabel='number of goal')
ax[1].plot(mean_year, color='green',label='Normal')
ax.flat[1].set(xlabel='month' ,ylabel='mean goal')

ax[2].plot(sum_month, color='green',label='Normal')
ax.flat[2].set(xlabel='year', ylabel='number of goal')

ax[3].plot(mean_month, color='green',label='Normal')
ax.flat[3].set(xlabel='month' , ylabel='mean goal')
```

```python

fig, ax = plt.subplots(2 , figsize=(18,15))
fig.suptitle("comparing Score and adj_score")
ax[0].plot(spi_matches["score1"], color='green',label='score1')
ax[0].plot(spi_matches["adj_score1"], color='yellow',label='adj_score1')
ax.flat[0].set(xlabel='game',ylabel='value')
ax[0].set_xlim([0, 150]) , ax[0].set_ylim([0, 7])
ax[0].legend(fontsize=12)

ax[1].plot(spi_matches["score2"], color='green',label='score2')
ax[1].plot(spi_matches["adj_score2"], color='yellow',label='adj_score2')
ax.flat[1].set(xlabel='game',ylabel='value')
ax[1].set_xlim([0, 150]) , ax[1].set_ylim([0, 5])
ax[1].legend(fontsize=12)
```

```python
Home_Team = spi_matches.groupby("team1")["score1"].sum().reset_index()
Home_Team = Home_Team.sort_values(by="score1",ascending= False)
plt.figure(figsize=(15,10))
ax = sns.barplot(y="team1",x="score1",
                 data=Home_Team[:20],palette="rainbow",
                 linewidth = 1,edgecolor = "k"*20)
plt.ylabel('')
plt.title("Top Home Teams with highest goals")
for i,j in enumerate(Home_Team["score1"][:20]):
    ax.text(1,i,j,weight = "bold")
```

```python
Away_Team = spi_matches.groupby("team2")["score2"].sum().reset_index()
Away_Team = Away_Team.sort_values(by="score2",ascending= False)
plt.figure(figsize=(15,10))
ax = sns.barplot(y="team2",x="score2",
                 data=Away_Team[:20],palette="rainbow",
                 linewidth = 1,edgecolor = "k"*20)
plt.ylabel('')
plt.title("Top Away Teams with highest goals")
for i,j in enumerate(Away_Team["score2"][:20]):
    ax.text(1,i,j,weight = "bold")
```
